```
install.packages("devtools")
devtools::install_github('cttobin/ggthemr')

library(ggthemr)
ggthemr("dust")
```

```
df <- data.frame(gp = factor(rep(letters[1:3], each = 10)),y = rnorm(30))
ggplot(df, aes(x = gp, y = y))+geom_boxplot()#畫出盒狀圖, x軸為gp, y軸為y
```

```
ggplot(df,aes(x = gp, y = y))+geom_line()+facet_grid(gp~.)#劃出折線圖(?), x軸為gp, y軸為y
```

```
ggplot()+geom_bar(data=NBA1314,aes(x=GamesPlayed,y=TotalMinutesPlayed),stat = "identity")#比較GamesPlayed(x座標)和TotalMinutesPlayed(y座標), 用bar chart表示
```

```
ggplot()+geom_point(data=NBA1314,aes(x=GamesPlayed,y=TotalMinutesPlayed),stat = "identity")#比較GamesPlayed(x座標)和TotalMinutesPlayed(y座標),用點狀分布表示
```
